```python
import duckdb

conn = duckdb.connect()
conn.execute("INSTALL httpfs; LOAD httpfs;")

# Configuración correcta para MinIO local
conn.execute("SET s3_endpoint='localhost:9000';")
conn.execute("SET s3_use_ssl=false;")
conn.execute("SET s3_url_style='path';")  # ← esta es la línea que faltaba
conn.execute("SET s3_access_key_id='minio';")
conn.execute("SET s3_secret_access_key='minio123';")

query = """
SELECT * FROM parquet_scan(
    's3://silver/wow_raid_events/v1/raid_id=raid001/event_date=*/part-*.parquet'
)
LIMIT 1
"""

result = conn.execute(query).df()
print(result.columns.tolist())
print(result.dtypes)
```
